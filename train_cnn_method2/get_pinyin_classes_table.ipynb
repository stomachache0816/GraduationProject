```python
import os

import requests
from bs4 import BeautifulSoup
import pandas as pd
from io import StringIO
```

```python
url = 'https://www.ifreesite.com/phonetic/phonetic.htm'
web = requests.get(url)
soup = BeautifulSoup(web.text, "html.parser")
# soup
```

```python
table_html_list = soup.select('table')
# table_html_list
```

```python
len(table_html_list)
```

```python
table_html = table_html_list[5]
# table_html
```

```python
table_content = pd.read_html(StringIO(table_html.prettify()))
# table_content
```

```python
type(table_content)
```

```python
len(table_content)
```

```python
df_table = table_content[0]
df_table
```
