```python
import pandas as pd
import re
import numpy as np
```

```python
df = pd.read_csv(filepath_or_buffer=f"..\\pinyin_classes_table.csv")
df
```

```python
df.loc[0, "Unnamed: 23"] = "純聲符"
df.loc[df.index[-1], "韻符↓  聲符→.1"] = "純韻符"
df.loc[0, "韻符↓  聲符→.1"] = np.nan
df = df.drop(columns=["韻符↓  聲符→"])
df.rename(columns={"韻符↓  聲符→.1": "韻符↓  聲符→", "Unnamed: 23": "呼音"}, inplace=True)
df
```

```python
for i in range(1, df.index[-1]):
    new_value = ""
    for j in range(len(df["呼音"][i])):
        if df["呼音"][i][j] != " ":
            new_value += df["呼音"][i][j]
    df.loc[i, "呼音"] = new_value

df
```

```python
pinyin_data_list = []

columns = df.columns[:-1]
for row_idx in range(0, df.index[-1]):
    for column_name in columns:
        if pd.notna(df.loc[row_idx, column_name]):
            datas = re.split(r'\s+', df.loc[row_idx, column_name])
            dic = {
                "bopomofo": datas[0],
                "pinyin": datas[2],
                "hu_yin": df.loc[row_idx, "呼音"],
                "she2_yin": df.loc[df.index[-1], column_name],
            }
            pinyin_data_list.append(dic)
            
pinyin_data_list
```

```python
hu_yin_label_dict = {
    "純聲符": 0, "開口呼": 1, "齊齒呼": 2, "合口呼": 3 , "撮口呼": 4
}

she2_yin_label_dict = {
    "純韻符": 0, "口唇音": 1, "舌尖音": 2, "舌根音": 3 , "舌面音": 4, "翹舌音": 5, "平舌音": 6
}

class_label_dic: dict = {}
label = 0

for pinyin_data in pinyin_data_list:
    hu_yin = pinyin_data["hu_yin"]
    she2_yin = pinyin_data["she2_yin"]
    key = (hu_yin, she2_yin)
    if key not in class_label_dic:
        class_label_dic[key] = label
        label += 1

print(f"len(class_label_dic): {len(class_label_dic)}")
class_label_dic
```

```python
hu_yin_label_dict = {
    "純聲符": 0, "開口呼": 1, "齊齒呼": 2, "合口呼": 3 , "撮口呼": 4
}

she2_yin_label_dict = {
    "純韻符": 0, "口唇音": 1, "舌尖音": 2, "舌根音": 3 , "舌面音": 4, "翹舌音": 5, "平舌音": 6
}

for i in range(len(pinyin_data_list)):
    hu_yin = pinyin_data_list[i]["hu_yin"]
    she2_yin = pinyin_data_list[i]["she2_yin"]
    
    pinyin_data_list[i]["hu_yin_label"] = hu_yin_label_dict[hu_yin]
    pinyin_data_list[i]["she2_yin_label"] = she2_yin_label_dict[she2_yin]
    pinyin_data_list[i]["class_label"] = class_label_dic[(hu_yin, she2_yin)]
    
pinyin_data_list
```

```python
df = pd.DataFrame(pinyin_data_list)
df
```

```python
df.to_csv(path_or_buf=f".\\pinyin_classes_layer_1_table.csv", index=False, encoding='utf-8')
```
