```python
import numpy as np
import pandas as pd
```

```python
edu_df = pd.read_csv(".\\edu_pinyin_label_table.csv")
edu_df
```

```python
class_df = pd.read_csv(".\\class_pinyin_label_table.csv")
class_df
```

```python
class_df[~class_df["bopomofo"].isin(edu_df["bopomofo_non_accent"])]
```

```python
bopomofo_list = list(class_df["bopomofo"])
for i in range(len(bopomofo_list)):
    bopomofo_list[i] = bopomofo_list[i].replace("一", "ㄧ")
class_df["bopomofo"] = bopomofo_list
class_df[~class_df["bopomofo"].isin(edu_df["bopomofo_non_accent"])]
```

```python
class_df[~class_df["pinyin"].isin(edu_df["pinyin_non_accent"])]
```

```python
pinyin_list = list(class_df["pinyin"].fillna(value="nan"))
for i in range(len(pinyin_list)):
    pinyin_list[i] = pinyin_list[i].replace("ê", "eh").replace("ɑ", "a").replace("ü", "v").replace("ɡ", "g")
class_df["pinyin"] = pinyin_list
class_df[~class_df["pinyin"].isin(edu_df["pinyin_non_accent"])]
```

```python
edu_df = edu_df.fillna(value="nan")
class_df[~class_df["pinyin"].isin(edu_df["pinyin_non_accent"])]
```

```python
edu_df[~edu_df["bopomofo_non_accent"].isin(class_df["bopomofo"])]
```

```python
class_df.merge(edu_df, left_on=["bopomofo", "pinyin"], right_on=["bopomofo_non_accent", "pinyin_non_accent"])
```

```python
edu_df.to_csv(".\\corrected_edu_pinyin_label_table.csv", encoding="utf-8", index=False)
class_df.to_csv(".\\corrected_class_df_pinyin_label_table.csv", encoding="utf-8", index=False)
```
